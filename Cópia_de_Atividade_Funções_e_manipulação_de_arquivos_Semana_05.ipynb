Escreva uma função chamada verificar_paridade que recebe um número como parâmetro e retorna True se for par e False se for ímpar.

```python
def verificar_pariedade(numero):
  if numero % 2 == 0:
    return True
  else:
    return False

print(verificar_pariedade(2))
```

Escreva uma função chamada verificar_palindromo que recebe uma string como parâmetro e retorna True se for um palíndromo (uma palavra que é igual se lida da esquerda para a direita e vice-versa) e False caso contrário.

```python
def verificar_palindromo(palavra):
  if palavra == palavra[::-1]:
    return True
  else:
    return False

print(verificar_palindromo('arara'))
```

##Desafio listas e funções:

Crie uma função que representa o jogo batalha naval.

Dentro da função crie uma estrutura bidimensional utilizando listas com sublistas para representar um tabuleiro (1 lista com 20 elementos e cada elemento é uma lista de 20 elementos, tabuleiro 20x20). Cada posição irá armazenar 1 valor numérico que significa:
- 0 - Água
- 1 - Navio

Para cada posição escolha esses valores aleatoriamente, respeitando a regra de que não podem existir mais de 20 navios no tabuleiro.

Após os valores serem distribuídos, o programa deve pedir ao usuário uma posição do tabuleiro e informar se ele acertou um navio ou água e repetir o procedimento até que o usuário derrote todos os navios ou chegue ao limite de 35 tentativas.



```python
import random


#Printar o tabuleiro
def printar_tabuleiro(tabuleiro):
  print("Tabuleiro:")
  for row in tabuleiro:
    print(row)

# Criação do tabuleiro
def criar_tabuleiro():
    tabuleiro = [[0 for _ in range(20)] for _ in range(20)]
    return tabuleiro

# Colocação dos navios no tabuleiro
def colocar_navios(tabuleiro):
  navios_colocados = 0

  while navios_colocados < 20:
      x = random.randint(0, 19)
      y = random.randint(0, 19)
      if tabuleiro[x][y] == 0:
          tabuleiro[x][y] = 1
          navios_colocados += 1
  return tabuleiro

# Jogo
def jogo(tabuleiro):
  tentativas = 0


  while tentativas < 35:
      print("Tentativa", tentativas + 1)
      x = int(input("Digite a coordenada X (1-20): "))-1

      while x <0 or x > 19:
          print("Coordenadas inválidas. Tente novamente.")
          x = int(input("Digite a coordenada X (1-20): "))-1

      y = int(input("Digite a coordenada Y (1-20): "))-1
      while y <0 or y > 19:
          print("Coordenadas inválidas. Tente novamente.")
          y = int(input("Digite a coordenada Y (1-20): "))-1


      print (f'{x},{y}')
      print (f'{tabuleiro[x][y]}')

      if tabuleiro[x][y] == 1:
        print("Você acertou um navio!")
        tabuleiro[x][y] = 0
        if sum(sum(row) for row in tabuleiro) == 0:
            print("Parabéns, você afundou todos os navios!")
            break
      else:
         print("Você errou!")
      tentativas += 1
  else:
    print("Você excedeu o limite de tentativas. Game over.")

  return 'teste'

def principal():
  tabuleiro = criar_tabuleiro()
  printar_tabuleiro(tabuleiro)
  tabuleiro = colocar_navios(tabuleiro)
  printar_tabuleiro(tabuleiro)
  jogo(tabuleiro)

principal()
```

Modifique o programa anterior para exibir as seguintes estatísticas.
- Acertos em água
- Acertos em Navios
- Porcentagem de acertos em água
- Porcentagem de acertos em Navios
- Acertos ininterruptos (maior quantidade de acertos em sequência)

```python
import random

#Printar o tabuleiro
def printar_tabuleiro(tabuleiro):
  print("Tabuleiro:")
  for row in tabuleiro:
    print(row)

# Criação do tabuleiro
def criar_tabuleiro():
    tabuleiro = [[0 for _ in range(20)] for _ in range(20)]
    return tabuleiro

# Colocação dos navios no tabuleiro
def colocar_navios(tabuleiro):
  navios_colocados = 0

  while navios_colocados < 20:
      x = random.randint(0, 19)
      y = random.randint(0, 19)
      if tabuleiro[x][y] == 0:
          tabuleiro[x][y] = 1
          navios_colocados += 1
  return tabuleiro

# Jogo
def jogo(tabuleiro):
  tentativas = 0
  qtde_acertos = 0
  qtde_erros = 0
  max_sequencia = 0
  sequencia = 0
  while tentativas < 35:
      print("Tentativa", tentativas + 1)
      x = int(input("Digite a coordenada X (1-20): "))-1
      while x <0 or x > 19:
          print("Coordenadas inválidas. Tente novamente.")
          x = int(input("Digite a coordenada X (1-20): "))-1

      y = int(input("Digite a coordenada Y (1-20): "))-1
      while y <0 or y > 19:
          print("Coordenadas inválidas. Tente novamente.")
          y = int(input("Digite a coordenada Y (1-20): "))-1

      print (f'{x},{y}')
      print (f'{tabuleiro[x][y]}')

      if tabuleiro[x][y] == 1:
        print("Você acertou um navio!")
        tabuleiro[x][y] = 0
        qtde_acertos += 1
        sequencia += 1
        if sequencia > max_sequencia:
            max_sequencia = sequencia
        if sum(sum(row) for row in tabuleiro) == 0:
            print("Parabéns, você afundou todos os navios!")
            break
      else:
         print("Você errou!")
         qtde_erros += 1
         sequencia = 0
      tentativas += 1
  else:
    print("Você excedeu o limite de tentativas. Game over.")

  resultados(qtde_erros,qtde_acertos,tentativas,max_sequencia)

  return 'Fim de jogo!'

def resultados(g_qtde_erros,g_qtde_acertos,g_tentativas,g_max_sequencia):
  print("Resultados:")
  print("Acertos em água:", g_qtde_erros)
  print("Acertos em Navios:", g_qtde_acertos)
  print("Porcentagem de acertos em água:", (g_qtde_erros / g_tentativas) * 100)
  print("Porcentagem de acertos em Navios:", (g_qtde_acertos / g_tentativas) * 100)
  print("Acertos ininterruptos:", g_max_sequencia)

def principal():
  tabuleiro = criar_tabuleiro()
  printar_tabuleiro(tabuleiro)
  tabuleiro = colocar_navios(tabuleiro)
  printar_tabuleiro(tabuleiro)
  jogo(tabuleiro)

principal()
```
