# Nova seção

Qual a quantidade de pessoas nas seguintes faixas etárias
- 20 anos ou menos
- 21 a 35 anos
- 36 a 50 anos
- 51 a 65 anos
- maiores de 65 anos



```python
#Tentando dimunir o volume de dados
idade_1 = df_pesquisa.loc[df_pesquisa.Idade<=20]['Idade'].count()
idade_2 = df_pesquisa.loc[(df_pesquisa.Idade>=21) & (df_pesquisa.Idade<=35)]['Idade'].count()
idade_3 = df_pesquisa.loc[(df_pesquisa.Idade>=36) & (df_pesquisa.Idade<=50)]['Idade'].count()
idade_4 = df_pesquisa.loc[(df_pesquisa.Idade>=51) & (df_pesquisa.Idade<=65)]['Idade'].count()
idade_5 = df_pesquisa.loc[df_pesquisa.Idade>65]['Idade'].count()

print(f'Total de pessoas com 20 anos ou menos: {idade_1}')
print(f'Total de pessoas com 21 a 35 anos: {idade_2}')
print(f'Total de pessoas com 36 a 50 anos: {idade_3}')
print(f'Total de pessoas com 51 a 65 anos: {idade_4}')
print(f'Total de pessoas maiores de 65 anos: {idade_5}')

#Conferindo
print(f'Total de pessoas: {df_pesquisa.shape[0]}')
print(f'Total de pessoas localizadas:' , idade_1 + idade_2 + idade_3 + idade_4 + idade_5)
```

Qual a UF que concentra a maior renda?


```python
#Criei um data frame apenas com as colunas de uf e renda
df_rendauf = df_pesquisa[['UF', 'Renda']]

#Fiz a soma por agrupamento
df_rendauf = df_rendauf.groupby('UF').sum()

#achei a maior renda por uf
max_renda = df_rendauf['Renda'].max()

max_renda #22539969

#pesquisando a renda no dataframe
df_index = df_rendauf.loc[df_rendauf['Renda'] ==  max_renda][['Renda']]

df_index = pd.DataFrame(df_index)

df_index.reset_index(inplace=True)

print(f'A maior renda é da UF {df_index.UF.values} com a renda de {df_index.Renda.values} ')
```

Qual a renda média de pessoas pretas e brancas de homens e mulheres?


```python
#Reduzindo dados
df_renda_sexo_cor = df_pesquisa[['Renda','Sexo','Cor']]

#Agrupando
#df_renda_sexo_cor = df_renda_sexo_cor.groupby(['Sexo','Cor'])['Renda'].sum()
#Preciso agrupar um quantidade tbm, count e sum mesmo comando = agg
df_renda_sexo_cor = df_renda_sexo_cor.groupby(['Sexo','Cor']).agg({'Renda': ['sum','count']})

df_renda_sexo_cor = pd.DataFrame(df_renda_sexo_cor)

#Adicionando coluna Media usando df_renda_sexo_cor.Sum / df_renda_sexo_cor.Count
df_renda_sexo_cor.columns = ['sum','count']

df_renda_sexo_cor["Media"] =  df_renda_sexo_cor[('sum')] / df_renda_sexo_cor[('count')]

df_renda_sexo_cor


#Filtrando por cor 2 - Branca e 4 -Pretas
#0	Masculino 1	Feminino


#Tive que resetar o index para filtrar Sexo como atributo
df_renda_sexo_cor.reset_index(inplace=True)

#Mulheres Negras
print('Média de renda salarial para Mulheres Negras',df_renda_sexo_cor.loc[(df_renda_sexo_cor.Sexo ==1)  & (df_renda_sexo_cor.Cor == 4)]['Media'].values)

#Mulheres Brancas
print('Média de renda salarial para Mulheres Brancas',df_renda_sexo_cor.loc[(df_renda_sexo_cor.Sexo ==1)  & (df_renda_sexo_cor.Cor == 2)]['Media'].values)

#Homens Negros
print('Média de renda salarial para Homens Negros',df_renda_sexo_cor.loc[(df_renda_sexo_cor.Sexo ==0)  & (df_renda_sexo_cor.Cor == 4)]['Media'].values)

#Homens Brancos
print('Média de renda salarial para Homens Brancos',df_renda_sexo_cor.loc[(df_renda_sexo_cor.Sexo ==0)  & (df_renda_sexo_cor.Cor == 2)]['Media'].values)

#renda_f_n = df_renda_sexo_cor.loc[(df_renda_sexo_cor.Sexo ==1)



```

 Qual a renda média concentrada em cada faixa etária?




```python
#Tentando dimunir o volume de dados

df_idade_1 = df_pesquisa.loc[df_pesquisa.Idade<=20].agg({'Renda': ['sum', 'count']})
df_idade_2 = df_pesquisa.loc[(df_pesquisa.Idade>=21) & (df_pesquisa.Idade<=35)].agg({'Renda': ['sum', 'count']})
df_idade_3 = df_pesquisa.loc[(df_pesquisa.Idade>=36) & (df_pesquisa.Idade<=50)].agg({'Renda': ['sum', 'count']})
df_idade_4 = df_pesquisa.loc[(df_pesquisa.Idade>=51) & (df_pesquisa.Idade<=65)].agg({'Renda': ['sum', 'count']})
df_idade_5 = df_pesquisa.loc[df_pesquisa.Idade>65].agg({'Renda': ['sum', 'count']})

media_idade_1 = df_idade_1.Renda['sum'] / df_idade_1.Renda['count']
media_idade_2 = df_idade_2.Renda['sum'] / df_idade_2.Renda['count']
media_idade_3 = df_idade_3.Renda['sum'] / df_idade_3.Renda['count']
media_idade_4 = df_idade_4.Renda['sum'] / df_idade_4.Renda['count']
media_idade_5 = df_idade_5.Renda['sum'] / df_idade_5.Renda['count']

print(f'Média de renda de pessoas com 20 anos ou menos: {media_idade_1}')
print(f'Média de renda de pessoas com 21 a 35 anos: {media_idade_2}')
print(f'Média de renda de pessoas com 36 a 50 anos: {media_idade_3}')
print(f'Média de renda de pessoas com 51 a 65 anos: {media_idade_4}')
print(f'Média de renda de pessoas maiores de 65 anos: {media_idade_5}')
```
